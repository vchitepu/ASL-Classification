```python
import warnings
warnings.simplefilter(action='ignore', category=DeprecationWarning)
warnings.simplefilter(action='ignore', category=FutureWarning)

import os
import cv2
import random
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn import model_selection
```

```python
DATADIR = 'Team_Data'
DATADIR1 = 'Team_Data_P1'
NAMES = ['Alex','Vinay','Delaney']
CATEGORIES = ['A','B','C','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y']

testing = "delaney"
testing = testing.lower()
if testing in NAMES:
    training = [x for x in NAMES if x != testing]
training
```

```python
def create_data(testing = None, question = None, method='training', res = (35,35)):
    
    training_data = []
    testing_data = []
    if question is 'part1':
        d = DATADIR1
        testing = testing.lower()
        if testing in NAMES:
            training = [x for x in NAMES if x != testing]
        else:
            print("Please enter who's data should be used for testing")
            return
        for name in training:
            for category in CATEGORIES:
                path = os.path.join(d,name)
                path = os.path.join(path,category)
                class_num = CATEGORIES.index(category)
                if method.lower() == 'training':
                    for img in os.listdir(path):
                        try:
                            
                            img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                            img_array = cv2.resize(img_array, res)
                            img_array = img_array.flatten()
                            training_data.append([img_array, class_num])

                            training_data = [x for x in training_data if x[0] is not None]
                            training_data = [x for x in training_data if len(x[0]) == res[0]*res[1]]
                            

                        except Exception as e:
                            pass
                elif method.lower() == 'images':
                    for img in os.listdir(path):
                        try:
                            img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                            img_array = cv2.resize(img_array, res)
                            training_data.append(img_array)

                        except Exception as e:
                            pass
        

        for category in CATEGORIES:
            path = os.path.join(d,testing)
            path = os.path.join(path,category)
            class_num = CATEGORIES.index(category)
            if method.lower() == 'training':
                for img in os.listdir(path):
                    try:
                        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                        img_array = cv2.resize(img_array, res)
                        img_array = img_array.flatten()
                        testing_data.append([img_array, class_num])

                        testing_data = [x for x in testing_data if x[0] is not None]
                        testing_data = [x for x in testing_data if len(x[0]) == res[0]*res[1]]

                    except Exception as e:
                        pass
            elif method.lower() == 'images':
                for img in os.listdir(path):
                    try:
                        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                        img_array = cv2.resize(img_array, res)
                        testing_data.append(img_array)
                        
                    except Exception as e:
                        pass
        
        return(training_data, testing_data)

                
            
    else:
        d = DATADIR
        for category in CATEGORIES:
            path = os.path.join(d, category)
            class_num = CATEGORIES.index(category)
            if method.lower() == 'training':
                for img in os.listdir(path):
                    try:
                        
                        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                        img_array = cv2.resize(img_array, res)
                        img_array = img_array.flatten()
                        training_data.append([img_array, class_num])

                        training_data = [x for x in training_data if x[0] is not None]
                        training_data = [x for x in training_data if len(x[0]) == res[0]*res[1]]

                    except Exception as e:
                        pass
            elif method.lower() == 'images':
                for img in os.listdir(path):
                    try:
                        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                        img_array = cv2.resize(img_array, (30,30))
                        training_data.append(img_array)

                    except Exception as e:
                        pass


        return training_data
```

# Part 1

```python
train, test = create_data(question = 'part1', testing = 'Delaney', method = 'training')

# Specify who you want to use as testing data with the testing parameter
```

```python
print(len(train))
print(len(test))
```

```python
# Formatting for use with classifier
X_train = []; y_train = []; X_test = []; y_test = []
for features, targets in train:
    X_train.append(features)
    y_train.append(targets)

for features, targets in test:
    X_test.append(features)
    y_test.append(targets)

X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

### Using SVM

```python
# Normalize
X_train = tf.keras.utils.normalize(X_train)
X_test = tf.keras.utils.normalize(X_test)
```

```python
# Import Classifer
from sklearn import svm

# Initialize classifer
clf = svm.SVC(gamma = 0.0001, C = 100)

# Training
clf.fit(X_train, y_train)

# Testing
acc = clf.score(X_test, y_test)
```

```python
acc
```

### Using Neural Network

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(146, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(24, activation = tf.nn.softmax))

model.compile(optimizer = 'adagrad',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1500)
```

```python
val_loss, val_acc = model.evaluate(X_test, y_test)
print(val_loss, val_acc)
```

# Part 2

```python
training_data = create_data(method = 'training', res = (64,64))
```

```python
random.shuffle(training_data)
```

```python
X = []; y = []
for features, targets in training_data:
    X.append(features)
    y.append(targets)
             
X = np.array(X)
y = np.array(y)
```

```python
print(X.shape)
print(y.shape)
```

### Using SVM

```python
# Cross_Validation
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.5)

X_train = tf.keras.utils.normalize(X_train)
X_test = tf.keras.utils.normalize(X_test)
```

```python
# Import Classifer
from sklearn import svm

# Initialize classifer
clf = svm.SVC(gamma = 0.0001, C = 100)

# Training
clf.fit(X_train, y_train)

# Testing
acc = clf.score(X_test, y_test)
```

```python
acc
```

### Using Deep Neural Network

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(146, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(146, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(146, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(24, activation = tf.nn.softmax))

model.compile(optimizer = 'adagrad',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=400)
```

```python
val_loss, val_acc = model.evaluate(X_test, y_test)
print(val_loss, val_acc)
```
